# **7. Advanced Analysis**

> **Based on the streams data, can you identify any patterns among tracks with the same key or mode (Major vs. Minor)?**
* *Tracks with Major mode are more likely to get streams than tracks with Minor mode. The tracks within C# regardless of its mode has the most streams.*

```python
# Remove rows where the 'key' column has NaN values
# This ensures that only tracks with a defined musical key are considered
nan_key = df.dropna(subset='key')

# Create a multi-level index using 'track_name', 'key', and 'mode', then sort the DataFrame by 'streams' in descending order.
getting_tracks = nan_key.set_index(['track_name', 'key', 'mode']).sort_values(by='streams', ascending=False)

# Display the DataFrame with the new multi-index of track name, key, and mode.
getting_tracks

# Group the data by 'key' and 'mode', and sum the total streams for each combination of key and mode.
grouped_data = nan_key.groupby(['key', 'mode'])['streams'].sum().reset_index()

# Create a bar plot to show the total streams categorized by key and mode.
plt.figure(figsize=(14, 8))  # Adjust the size of the plot for better readability.
sns.barplot(data=grouped_data, x='key', y='streams', hue='mode', palette=['m', 'c'])
# 'hue' differentiates between different modes (major/minor) within the same key.

# Add a title and axis labels to describe the plot clearly.
plt.title('Total Streams by Key and Mode', fontsize=16)  # Set the title of the plot.
plt.xlabel('Musical Key', fontsize=14)  # Label the x-axis with 'Musical Key'.
plt.ylabel('Total Streams', fontsize=14)  # Label the y-axis with 'Total Streams'.

# Display the legend showing the mode (major or minor) in the upper right corner of the plot.
plt.legend(title='Mode', loc='upper right', fontsize=12)

# Ensure that all elements of the plot fit properly in the display.
plt.tight_layout()

# Display the plot
plt.show()
```

> **Do certain genres or artists consistently appear in more playlists or charts? Perform an analysis to compare the most frequently appearing artists in playlists or charts.**
* *Based on the plot generated, The Weeknd appears most frequently in playlists, and Taylor Swift appears more frequently in charts. Overall, The Weeknd has the most appearances in playlists and charts in total.*

```python
# Define the columns for playlists and charts to analyze the number of appearances across different platforms.
playlist_cols = ['in_spotify_playlists', 'in_apple_playlists', 'in_deezer_playlists']
chart_cols = ['in_spotify_charts', 'in_apple_charts', 'in_deezer_charts', 'in_shazam_charts']

# Convert the playlist and chart columns to numeric values, replacing non-numeric entries with NaN.
df[playlist_cols + chart_cols] = df[playlist_cols + chart_cols].apply(pd.to_numeric, errors='coerce')

# Calculate the total number of appearances in playlists by summing values across the playlist columns for each track.
df['total_playlist_appearances'] = df[playlist_cols].sum(axis=1)

# Calculate the total number of appearances in charts by summing values across the chart columns for each track.
df['total_chart_appearances'] = df[chart_cols].sum(axis=1)

# Group the data by artist name and calculate the sum of their total playlist and chart appearances.
artist_appearances = df.groupby('artist(s)_name', as_index=False)[['total_playlist_appearances', 'total_chart_appearances']].sum()

# Calculate the overall appearances for each artist by adding their playlist and chart appearances.
artist_appearances['total_appearances'] = artist_appearances['total_playlist_appearances'] + artist_appearances['total_chart_appearances']

# Identify the top 10 artists based on their total appearances across playlists and charts.
top_artists = artist_appearances.nlargest(10, 'total_appearances')

# Reshape the data for visualization using the melt function to create separate rows for each type of appearance (total, playlist, chart).
top_artists_melted = top_artists.melt(id_vars='artist(s)_name',
                                      value_vars=['total_appearances', 'total_playlist_appearances', 'total_chart_appearances'],
                                      var_name='Appearance Type',
                                      value_name='Count')

# Create a bar plot showing the top 10 artists with their total appearances, playlist appearances, and chart appearances.
plt.figure(figsize=(14, 8))  # Adjust the size of the plot for better readability.
sns.barplot(data=top_artists_melted, x='artist(s)_name', y='Count', hue='Appearance Type', palette=['m', 'c', 'r'])
# 'hue' differentiates between total appearances, playlist appearances, and chart appearances.

# Add the title and axis labels for context.
plt.title('Top 10 Artists by Playlist and Chart Appearances', fontsize=16)
plt.xlabel('Artist Name', fontsize=14)  # Label the x-axis as 'Artist Name'.
plt.ylabel('Appearance Count', fontsize=14)  # Label the y-axis as 'Appearance Count'.

# Display a legend to indicate the appearance type represented by each color.
plt.legend(title='Appearance Type', loc='upper right', fontsize=12)

# Rotate the x-axis labels slightly for improved readability.
plt.xticks(rotation=45)

# Adjust the layout to ensure that all plot elements are well-spaced and visible.
plt.tight_layout()

# Display the plot.
plt.show()
```
