# **2. Basic Descriptive Statistics**

```python
## Basic Descriptive Statistics

# Summary of Statistics
print("Summary of Statistics:")
df[['streams', 'released_day', 'released_month', 'released_year', 'bpm', 'danceability_%']].describe()      # Generates descriptive statistics for the selected columns
```

**What are the mean, median, and standard deviation of the streams column?**
*   *The mean of the streams column is 514137424.93907565, the median is 290530915.0, and the standard deviation is 566856949.038883.*

```python
# Mean, Median, and Standard Deviation of the Streams Column

df['streams'] = pd.to_numeric(df['streams'], errors='coerce')       # Converts 'streams' column to numeric, if it's not already

# Mean
mean = df['streams'].mean()
print(f"Mean: {mean}")

# Meadian
median = df['streams'].median()
print(f"Median: {median}")

# Standard Deviation
std = df['streams'].std()
print(f"Standard Deviation: {std}")
```

# **4. Temporal Trends**

**Analyze the trends in the number of tracks released over time. Plot the number of tracks released per year.**
* *As seen in the plot attached below, there was a spike in the number of songs released during the 2020s.*

```python
# Analyze Trends in Tracks Released Over Time (Plotting by Year)

## Count the number of tracks released per year
tracks_per_year = df['released_year'].value_counts().sort_index()

## Plot the number of tracks released over time (yearly)
plt.plot(tracks_per_year.index, tracks_per_year.values, linestyle='-', color='m')
plt.title('Tracks Released Over Time')
plt.xlabel('Year')
plt.ylabel('Number of Tracks')
plt.grid(True)
plt.show()
```

# **5. Genre and Music Characteristics**

**Examine the correlation between streams and musical attributes like bpm, danceability_%, and energy_%. Which attributes seem to influence streams the most?**
* *danceability_% has a negative correlation with coefficient farthest from 0, with -0.11. This means that it influences streams the most but in the opposite direction.*

```python
# correlation between streams and musical attributes like bpm, danceability_%, and energy_%.
df['streams'] = pd.to_numeric(df.streams, errors='coerce')
corr = df[['streams','bpm','danceability_%','energy_%','valence_%','acousticness_%']].corr()

# for visualization
plt.figure(figsize= (5,6))
sns.set(font_scale=1)
sns.heatmap(corr.iloc[[0],1:], annot=True, cmap='coolwarm', linewidths=0.5)
plt.show()
```

# **7. Advanced Analysis**

**Based on the streams data, can you identify any patterns among tracks with the same key or mode (Major vs. Minor)?**

```python
## Identify any patterns among tracks with the same key or mode (Major vs. Minor)

# Group by 'key' and 'mode', then calculate the average streams
key_mode_group = df.groupby(['key', 'mode'])['streams'].mean().reset_index()

# Plot the results
plt.figure(figsize=(12,6))
sns.barplot(x='key', y='streams', hue='mode', data=key_mode_group, palette=['m','c'])
plt.title('Average Streams by Key and Mode')
plt.ylabel('Average Streams')
plt.show()
```

**Do certain genres or artists consistently appear in more playlists or charts? Perform an analysis to compare the most frequently appearing artists in playlists or charts.**
* *Based on the plots generated, Taylor Swift appeared most frequently in both Spotify Charts and Apple Playlists, and The Weeknd appeared most frequently in Spotify Playlists.*

```python
## Perform an analysis to compare the most frequently appearing artists in playlists or charts

# Group by artist and count the number of appearances in each platform
artist_platform_count = df.groupby('artist(s)_name')[['in_spotify_playlists', 'in_spotify_charts', 'in_apple_playlists']].sum().reset_index()

# Sort the artists based on appearances in each platform separately
top_spotify_playlists = artist_platform_count.sort_values(by='in_spotify_playlists', ascending=False).head(10)
top_spotify_charts = artist_platform_count.sort_values(by='in_spotify_charts', ascending=False).head(10)
top_apple_playlists = artist_platform_count.sort_values(by='in_apple_playlists', ascending=False).head(10)

# Plot Top 10 Artists in Spotify Playlists
plt.figure(figsize=(12, 6))
sns.barplot(x='artist(s)_name', y='in_spotify_playlists', data=top_spotify_playlists, color='g')
plt.title('Top 10 Artists by Spotify Playlists Appearances')
plt.xlabel('Artist')
plt.ylabel('Spotify Playlists Appearances')
plt.xticks(rotation=45)
plt.show()

# Plot Top 10 Artists in Spotify Charts
plt.figure(figsize=(12, 6))
sns.barplot(x='artist(s)_name', y='in_spotify_charts', data=top_spotify_charts, color='k')
plt.title('Top 10 Artists by Spotify Charts Appearances')
plt.xlabel('Artist')
plt.ylabel('Spotify Charts Appearances')
plt.xticks(rotation=45)
plt.show()

# Plot Top 10 Artists in Apple Playlists
plt.figure(figsize=(12, 6))
sns.barplot(x='artist(s)_name', y='in_apple_playlists', data=top_apple_playlists, color='r')
plt.title('Top 10 Artists by Apple Playlists Appearances')
plt.xlabel('Artist')
plt.ylabel('Apple Playlists Appearances')
plt.xticks(rotation=45)
plt.show()
```

```python
# Group by artist and count the number of appearances in playlists and charts
artist_playlist_count = df.groupby('artist(s)_name')[['in_spotify_playlists', 'in_spotify_charts', 'in_apple_playlists']].sum().reset_index()

# Sort the artists based on total playlist/chart appearances
artist_playlist_count['total_playlists'] = artist_playlist_count[['in_spotify_playlists', 'in_spotify_charts', 'in_apple_playlists']].sum(axis=1)
top_artists = artist_playlist_count.sort_values(by='total_playlists', ascending=False).head(10)

# Plot the top 10 artists appearing in playlists and charts
plt.figure(figsize=(12,6))
sns.barplot(x='artist(s)_name', y='total_playlists', data=top_artists, color='m')
plt.title('Top 10 Artists by Playlist and Chart Appearances')
plt.ylabel('Total Appearances')
plt.xticks(rotation=45)
plt.show()
```
